#### Definicija tenzorskog proizvoda

*Tenzorski proizvod* $V \otimes W$ vektorskih prostora $V$ i $W$ nad istim poljem skalara, sa bazama $\{\ket{v_1}, \ldots, \ket{v_n}\}$ i $\{\ket{w_1}, \ldots, \ket{w_n}\}$ je vektorski prostor dimenzije $n \cdot m$ Äiji su bazni vektori $v_i \otimes w_j$, za sve $1 \leq i \leq n$ i $1 \leq j \leq m$, gde je $\otimes$ operacija koja zadovoljava sledeÄ‡e aksiome:
\begin{align*}
    (\ket{v_1} + \ket{v_2}) \otimes \ket{w} = (\ket{v_1} \otimes \ket{w}) + (\ket{v_2} \otimes \ket{w}), \
    \ket{v} \otimes (\ket{w_1} + \ket{w_2}) = (\ket{v} \otimes \ket{w_1}) + (\ket{v} \otimes \ket{w}), \
    (a \cdot  \ket{v}) \otimes w = \ket{v} \otimes (a \cdot \ket{w}) = a (\ket{v} \otimes \ket{w}).
\end{align*}

#### Napomena:

NajÄeÅ¡Ä‡e se oznaka $\otimes$ izostavlja, osim kada moÅ¾e doÄ‡i do zabune. Tako moÅ¾emo pisati $\ket{0} \ket{0}$ umesto $\ket{0} \otimes \ket{0}$. Intuicija iza tenzorskog proizvoda je da se omoguÄ‡i definisanje niza kubita. MoÅ¾e se pokazati da je tenzorski proizvod dva Hilbertova prostora takoÄ‘e Hilbertov prostor. ### Kvantno merenje sistema sa viÅ¡e kubita

Projektivno kvantno merenje se moÅ¾e definisati na sliÄan naÄin kao i kod jednog kubita. Na primeru Ä‡emo pokazati kako radi takva vrsta merenja.

#### Primer:

Razmotrimo standardni Hilbertov prostor $\mathbb{C}^2$ sa baznim vektorima $\ket{0}$ i $\ket{1}$. ReÅ¡enje je:
    \begin{align*}
        \left(\bra{11}\left(\frac{1}{\sqrt{3}}\ket{00} + \frac{\sqrt{2}}{\sqrt{3}} \ket{11}\right)\right)^2 = \frac{2}{3}.

Osim merenja nad celim sistemom, moguÄ‡e je izvrÅ¡iti merenje nad pojedinaÄnim kubitima. Na primer, ako Å¾elimo da izvrÅ¡imo merenje nad prvim bitom sa bazom $\ket{0}$, $\ket{1}$, odgovarajuÄ‡e projekcije su $\ket{0}\bra{0} \otimes I$ i $\ket{1}\bra{1} \otimes I$.

Kubit $\ket{v}$ Ä‡e zavrÅ¡iti u stanju $\ket{00}$ sa verovatnoÄ‡om $\frac{1}{3}$.

### Kvantno sprezanje

*Kvantno sprezanje* (eng. *quantum entanglement*) je fenomen u kojem sistem od dva ili viÅ¡e kvantnih bitova (kubita) nije moguÄ‡e prikazati kao tenzorski proizvod pojedinaÄnih delova.

#### Definicija:

    \begin{align*}

    &\ket{\Psi^{+}} = \frac{1}{\sqrt{2}} (\ket{01} + \ket{10})\\

#### Primer:

    \begin{align*}
        \frac{1}{\sqrt{2}}(\ket{00} + \ket{11}) = (a_1 \ket{0} + a_2 \ket{1}) \otimes (b_1 \ket{0} + b_2 \ket{1})?

    \begin{align*}
        a_1 b_1 = \frac{1}{\sqrt{2}} \\

        a_2 b_1 = 0 \\
        a_2 b_2 = \frac{1}{\sqrt{2}}
    \end{align*}.

Ipak, sistem nema reÅ¡enja.

Na poÄetku ovog poglavlja, videli smo da viÅ¡e kubita formira sistem Äije se stanje opisuje tenzorskim proizvodom tih kubita. Dakle, sistem od dva ili viÅ¡e kubita moÅ¾e da se naÄ‘e u stanju u kojem nije moguÄ‡e jasno opisati delove koji ga Äine. U tom sluÄaju, delovi koji Äine taj sistem nisu formalno kubiti, na osnovu definicije kubita. Zbog jednostavnosti, *delove* koji Äine spregnut sistem sa viÅ¡e kubita Ä‡emo nazivati kubitima, iako njihovo stanje nije definisano. Merenjem nad jednim kubitom u bilo kojem od ovih bitova se menja stanje drugog kubita, odnosno ako merimo prvi bit kubita $\ket{\Phi^{+}}$, sa verovatnoÄ‡om $\frac{1}{2}$ Ä‡e izaÄ‡i $0$ kao rezultat i kubit Ä‡e zavrÅ¡iti u stanju $\ket{00}$.

### NemoguÄ‡nost kloniranja:

Jedna od vaÅ¾nih odlika kvantnih raÄunara je nemoguÄ‡nost *kloniranja* nepoznatih kubita. To znaÄi da ne postoji kvantno kolo $U$ takvo da vaÅ¾i:
\begin{align*}
    U \ket{v} \ket{0} = \ket{v}\ket{v}.
\end{align*}

Kvantno kolo $U$, ovako definisano, prima kubit $\ket{v}$ Äije kvantno stanje Å¾elimo da kopiramo i kubit $\ket{0}$ na koji Å¾elimo da upiÅ¡emo kvantno stanje od $\ket{v}$. NemoguÄ‡nost kloniranja nam govori da nije moguÄ‡e napraviti opÅ¡ti algoritam za kopiranje bilo kog kubita.

Ovaj rezultat pokazali su Vuters i Zurek 1982. Njihov dokaz je vrlo kratak i zasniva se na linearnosti operatora $U$ (podsetimo se, $U$ je unitarna linearna transformacija). Najpre, neka su $\ket{a}$ i $\ket{b}$ dva ortogonalna vektora. Tada je:
\begin{align*}
    U\ket{a}\ket{0} = \ket{a}\ket{a}, \\
    U\ket{b}\ket{0} = \ket{b}\ket{b}.
\end{align*}

Neka je $\ket{c} = \frac{1}{\sqrt{2}} (\ket{a} + \ket{b})$. Odavde sledi:
\begin{align*}
    U \ket{c} \ket{0} &= \ket{c} \ket{c} \\
    &= \frac{1}{2} (\ket{aa} + \ket{ab} + \ket{ba} + \ket{bb}) \neq \frac{1}{\sqrt{2}} (\ket{a}\ket{a} + \ket{b}\ket{b}) = \frac{1}{\sqrt{2}} (U\ket{a}\ket{0} + U\ket{b}\ket{0)} = U \ket{c}\ket{0}.
\end{align*}

Iako na prvi pogled nemoguÄ‡nost kloniranja izgleda kao velika mana kvantnih raÄunara, to je osnova mnogih algoritama iz oblasti kriptografije. Na sledeÄ‡em primeru Ä‡emo videti jedan kvantni protokol iz kriptografije.

#### Primer:

#### Q# implementacija protokola BB84

NaÅ¡a implementacija biÄ‡e samo simulacija rada algoritma. Ne postoji trenutno naÄin da pokrenemo algoritam u praksi, jer joÅ¡ ne postoji kvantni internet.

```python
%%qsharp

operation f1(q : Qubit, b : Bool) : Unit {
    if b {
        X(q);
    }
}

operation f2(q: Qubit, b: Bool) : Unit {
    if b {
        X(q);
    }
    H(q);
}

operation HadamardMeasurement(q : Qubit) : Result {
    H(q);
    return M(q);
}

operation generateRandomBit() : Bool {
    use q = Qubit();
    H(q);
    let m = M(q);
    // kada izaÄ‘emo iz funkcije, kubit se oslobaÄ‘a
    // da bi se uspeÅ¡no mogao osloboditi, potrebno je da naznaÄimo sistemu
    // da ne sadrÅ¾i nikakvu bitnu informaciju
    // zato moramo resetovati kubit na |0>
    Reset(q);
    if m == One {
        return true;
    }
    return false;
}

operation Alice(n : Int) : (Bool[], Bool[]) {
    // korak 1: generiÅ¡i n bitova nasumiÄno
    mutable bits = [false, size = n];
    // 0..n-1 je interval [0, n - 1]
    for i in 0..n-1 {
        bits[i] = generateRandomBit();
    }

    // korak 2: Alisa kpnstruiÅ¡e niz kubita
    use qubits = Qubit[n];
    mutable bases = [false, size = n];
    for i in 0..n-1 {
        let m = generateRandomBit();
        bases[i] = m;
        if m {
            f2(qubits[i], bits[i]);
        } else {
            f1(qubits[i], bits[i]);
        }
    }

    // korak 3: Alisa Å¡alje niz kubita Bobanu
    let bob = Bob1(qubits, n);

    let bobBases = Fst(bob);
    let bobBits = Snd(bob);

    // korak 6: Alisa sada govori Bobu koju je bazu koristila
    let matched = Bob2(bases, bobBases, n);

    // Alisa prepisuje saÄuvane bitove
    mutable m = 0;
    for i in 0..n-1 {
        if matched[i] {
            m += 1;
        }
    }

    mutable savedBits = [false, size = m];
    mutable savedBobBits = [false, size = m];
    mutable k = 0;
    for i in 0..n-1 {
        if matched[i] {
            // Alisa prepisuje saÄuvane bitove
            savedBits[k] = bits[i];
            // Bob prepisuje saÄuvane bitove
            savedBobBits[k] = bobBits[i];
            k += 1;
        }
    }

    // resetujemo kubite pre izlaska iz funkcije
    for i in 0..n-1 {
        Reset(qubits[i]);
    }

    // korak 7: Alisa proverava nasumiÄne bitove
    // radi jednostavnosti, neka bude da su to 2 bita sa kraja niza
    if m > 2 and savedBits[m - 1] == savedBobBits[m - 1] and savedBits[m - 2] == savedBobBits[m - 2] {
        return (savedBits, savedBobBits);
    } else {
        return ([false], [false]);
    }
}

operation Bob1(qubits : Qubit[], n : Int) : (Bool[], Bool[]) {
    // korak 4: Boban je primio niz kubita Äim je pokrenuta ova funkcija
    
    // korak 5: Boban generiÅ¡e niz bitova
    // true --> Adamarova baza
    // false --> kanonska
    mutable bobBases = [false, size = n];
    mutable bobBits = [false, size = n];
    for i in 0..n-1 {
        bobBases[i] = generateRandomBit();
        mutable m : Result = Zero;
        if bobBases[i] {
            m = HadamardMeasurement(qubits[i]);
        } else {
            m = M(qubits[i]);
        }
        if m == One {
            bobBits[i] = true;
        }
    }

    return (bobBases, bobBits);
}

function Bob2(bases : Bool[], bobBases : Bool[], n : Int) : Bool[] {
    // Boban proverava na kojim mestima je pogodio bazu
    mutable matched = [false, size = n];

    for i in 0..n-1 {
        if bases[i] == bobBases[i] {
            matched[i] = true;
        }
    }
    return matched;
}

let keys = Alice(20);

for i in 0..Length(Fst(keys))-1 {
    Message($"{Fst(keys)[i]}, {Snd(keys)[i]}");
}
```
